```
def agg_data(file_path, file_name, granularity='DAY', agg_func='mean', chunk_size=10**5):
  init_start_time = time.time()
  output = pd.DataFrame()
  group_cols = []
  if granularity == 'YEAR':
    output = pd.DataFrame(columns={'YEAR', 'PRICE'})
    group_cols = ['YEAR']
  elif granularity == 'MONTH':
    output = pd.DataFrame(columns={'YEAR', 'MONTH', 'PRICE'})
    group_cols = ['YEAR', 'MONTH']
  elif granularity == 'DAY':
    output = pd.DataFrame(columns={'YEAR', 'MONTH', 'DAY', 'PRICE'})
    group_cols = ['YEAR', 'MONTH', 'DAY']
  elif granularity == 'HOUR':
    output = pd.DataFrame(columns={'YEAR', 'MONTH', 'DAY', 'HOUR', 'PRICE'})
    group_cols = ['YEAR', 'MONTH', 'DAY', 'HOUR']
  elif granularity == 'MINUTE':
    output = pd.DataFrame(columns={'YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE', 'PRICE'})
    group_cols = ['YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE']
  elif granularity == 'MINUTE_VOL':
    output = pd.DataFrame(columns={'YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE', 'PRICE','VOLUME'})
    group_cols = ['YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE']
  else:
    raise Exception("Unsupported level of granularity")

  prev_step = pd.DataFrame(columns={'YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE', 'PRICE','VOL'})
  count = 0
  for chunk in pd.read_csv(file_path, chunksize=chunk_size, error_bad_lines=False):
    loop_start_time = time.time()
    chunk_df = pd.DataFrame({'YEAR': chunk.iloc[:,1].astype(str).str[:4].astype(int),
                     'MONTH': chunk.iloc[:,1].astype(str).str[4:6].astype(int),
                     'DAY': chunk.iloc[:,1].astype(str).str[6:].astype(int),
                     'HOUR': chunk.iloc[:,2].astype(str).str.split(":", expand=True)[0].astype(int),
                     'MINUTE': chunk.iloc[:,2].astype(str).str.split(":", expand=True)[1].astype(int),
                     'PRICE': chunk.iloc[:,3],
                     'VOL': chunk.iloc[:,4]})
    
    count += chunk_size
    chunk_df = chunk_df[(chunk_df['HOUR'] >= 9) & (chunk_df['HOUR'] < 16)]
    chunk_df = chunk_df[(chunk_df['HOUR'] > 9) | ((chunk_df['HOUR'] == 9) & (chunk_df['MINUTE'] >= 30))]

    chunk_df = prev_step.append(chunk_df,sort=False)
    chunk_df.index = range(len(chunk_df))

    mask = chunk_df[granularity] == chunk_df.loc[len(chunk_df) - 1, granularity]
    prev_step = chunk_df[mask]

    chunk_df = chunk_df[~ mask]

    chunk_df = chunk_df[output.columns]
    
    chunk_df = chunk_df.groupby(group_cols).agg(agg_func)
    chunk_df = chunk_df.reset_index()

    output = output.append(chunk_df, sort=False)
    print(output)
    print(str(count) + " rows processed")
    print(str(len(output)) + " " + granularity.lower() + "s aggregated")
    print("Overall Time: " + str(time.time() - init_start_time))
    print("Loop Time: " + str(time.time() - loop_start_time))
  
  output = output.groupby(group_cols).agg(agg_func)
  output = output.reset_index()
  output.index = range(len(output))
  output = output[group_cols]
  print(output)
  output.to_csv("/content/drive/My Drive/Stock Price Prediction Research/Data/" + file_name + ".csv", index=False)
```
